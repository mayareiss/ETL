```python
#Combine all data and remove index

rebags = [gucci_id,vuitton_id,chanel_id,prada_id,hermes_id,fendi_id,dior_id,valentino_id,louboutin_id,balenciaga_id]
rebag = pd.concat(rebags, axis=0, ignore_index=True)
rebag
```

```python
rebag["Type"] = ""
for index, row in rebag.iterrows():
    if('crossbody' in row["title"].lower()):
        row["Type"] = 'Crossbody'
    elif('tote'  in row["title"].lower()):
        row['Type'] = 'Tote'
    elif('clutch' in row["title"].lower()):
        row['Type'] = 'Clutch'
    elif('shoulder' in row["title"].lower()):
        row['Type'] = 'Shoulder'
    elif('backpack' in row["title"].lower()):
        row['Type'] = 'Backpack'
    elif('satchel' in row["title"].lower()):
        row['Type'] = 'Satchel'
    elif('mini' in row["title"].lower()):
        row['Type'] = 'Mini'    
    elif('cross-body' in row["title"].lower()):
        row['Type'] = 'Crossbody'    
    elif('messenger' in row["title"].lower()):
        row['Type'] = 'Messenger'  
    else:
        row['Type'] = 'Other'
rebag["Type"].value_counts()
```

```python
#Add material of bag

rebag['Material'] = ""
for index, row in rebag.iterrows():
    if('crocodile' in row['title'].lower()):
        row['Material'] = 'Reptile'
    elif('python'  in row['title'].lower()):
        row['Material'] = 'Reptile'
    elif('snake'  in row['title'].lower()):
        row['Material'] = 'Reptile'
    elif('canvas'  in row['title'].lower()):
        row['Material'] = 'Canvas'
    elif('lamb'  in row['title'].lower()):
        row['Material'] = 'Lamb'
    elif('silk' in row['title'].lower()):
        row['Material'] = 'Silk'  
    elif('suede' in row['title'].lower()):
        row['Material'] = 'Suede'  
    elif('alligator'  in row['title'].lower()):
        row['Material'] = 'Reptile'
    elif('lizard'  in row['title'].lower()):
        row['Material'] = 'Reptile'
    elif('calf'  in row['title'].lower()):
        row['Material'] = 'Calf'
    elif('leather'  in row['title'].lower()):
        row['Material'] = 'General Leather'
    else:
        row['Material'] = 'Other/Unknown'
rebag['Material'].value_counts()
```

```python
#Clean up Price column
rebag['price'] = rebag['price'].str.replace('$', '')
rebag['price'] = rebag['price'].str.replace(',', '')
rebag['price'] = rebag['price'].astype(float)
```
