```python
#Checking value
ebay['Status'].value_counts()
# ebay['Shipping Cost'].value_counts()
```

```python
#Add type of bag

ebay["Type"] = ""
for index, row in ebay.iterrows():
    if('crossbody' in row["Name"].lower()):
        row["Type"] = 'Crossbody'
    elif('tote'  in row["Name"].lower()):
        row['Type'] = 'Tote'
    elif('clutch' in row["Name"].lower()):
        row['Type'] = 'Clutch'
    elif('shoulder' in row["Name"].lower()):
        row['Type'] = 'Shoulder'
    elif('backpack' in row["Name"].lower()):
        row['Type'] = 'Backpack'
    elif('satchel' in row["Name"].lower()):
        row['Type'] = 'Satchel'
    elif('mini' in row["Name"].lower()):
        row['Type'] = 'Mini'    
    elif('cross-body' in row["Name"].lower()):
        row['Type'] = 'Crossbody'    
    elif('messenger' in row["Name"].lower()):
        row['Type'] = 'Messenger'  
    else:
        row['Type'] = 'Other'
ebay["Type"].value_counts()
```

```python
#Add material of bag

ebay['Material'] = ""
for index, row in ebay.iterrows():
    if('crocodile' in row['Name'].lower()):
        row['Material'] = 'Reptile'
    elif('python'  in row['Name'].lower()):
        row['Material'] = 'Reptile'
    elif('snake'  in row['Name'].lower()):
        row['Material'] = 'Reptile'
    elif('canvas'  in row['Name'].lower()):
        row['Material'] = 'Canvas'
    elif('lamb'  in row['Name'].lower()):
        row['Material'] = 'Lamb'
    elif('silk' in row['Name'].lower()):
        row['Material'] = 'Silk'  
    elif('suede' in row['Name'].lower()):
        row['Material'] = 'Suede'  
    elif('alligator'  in row['Name'].lower()):
        row['Material'] = 'Reptile'
    elif('lizard'  in row['Name'].lower()):
        row['Material'] = 'Reptile'
    elif('calf'  in row['Name'].lower()):
        row['Material'] = 'Calf'
#removing the next three because the total amount of each is neglible
#     elif('goat'  in row['Name'].lower()):
#         row['Material'] = 'Goat'
#     elif('sheep'  in row['Name'].lower()):
#         row['Material'] = 'Sheep'
#     elif('deer'  in row['Name'].lower()):
#         row['Material'] = 'Deer'
    elif('leather'  in row['Name'].lower()):
        row['Material'] = 'General Leather'
    else:
        row['Material'] = 'Other/Unknown'
ebay['Material'].value_counts()
```
